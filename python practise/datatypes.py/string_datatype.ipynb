```python
text = "National Aeronautics and Space Administration"

# Split the phrase into words
words = text.split()


words
```

```python
"".join([word[0].isupper for word in words])
```

```python
" ".join(words)
```

```python
# Take the first letter of each word and make it uppercase
acronym = "".join(word[0].upper() for word in words)

print(acronym)  # Output: NASA
```

```python
text = "National Aeronautics and Space Administration"

# List of words to ignore when forming the acronym
ignore_words = {"and", "of", "the", "for", "in", "on"}

# Split the phrase and take the first letter of each relevant word
acronym = "".join(
    word[0].upper() for word in text.split() 
    if word.lower() not in ignore_words
)

print(acronym)  # Output: NASA
```

```python
text = "National Aeronautics and Space Administration"
ignore = {"and", "of", "the"}  # words to skip

words = text.split()
acronym = "".join(word[0].upper() for word in words if word.lower() not in ignore)

print(acronym)  # Outputs: NASA
```
