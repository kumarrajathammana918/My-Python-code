```python
# Question 44: Check if string is a valid URL: "https://www.google.com"
import re
def is_validurl(url)
    pattern = r'^(http|https):\/\/([\w.-]+)(\.[\w.-]+)+([\/\w\.-]*)*\/?$'
    return bool(re.match(pattern, url))

```

```python
# Question 46: Convert "hello world" to Pig Latin
def pig_latin_word(sentence):
    y=sentence.split()
    x=[]
    for i in y:
       z=i[:1]+i[0]+"ay"
    x.append(z)
    return "".join(x)
text="hello world"
result=pig_latin_word(text)
print(result)



```
