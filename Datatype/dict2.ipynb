```python
# Question 40: Convert each generator to list
my_generators ={
    'gen1': (x for x in range(3)),      
    'gen2': (x * x for x in range(5, 8)) 
}


my_lists = {key: list(gen) for key, gen in my_generators.items()}

print(my_lists)

```
