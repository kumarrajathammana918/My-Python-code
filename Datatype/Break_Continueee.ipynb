```python
# Write a program that prints all numbers from 1 to 50, but skips (doesn't print)
#  numbers that are multiples of 3. Use the continue statement.

# Hint: Loop from 1 to 50. If i%3==0, use continue to skip to next iteration. Otherwise print i.


for i in range(1,51):
    if i%3==0:
        continue
    print(i)
```

```python
# Write a program that:

# Keeps asking the user to enter numbers
# Adds positive numbers to a sum
# Skips zero (continues to next iteration)
# Stops when a negative number is entered
# Prints the final sum
# Hint: Use while True loop. If num < 0, break. If num == 0, continue. Otherwise

sum=0
while True:
    num=int(input("Enter a number: "))
    if num<0:
        break
    if num==0:
        continue
    sum=sum+num
```

```python
# Problem 6.2: GCD Calculator (Medium)
# Write a program to find the Greatest Common Divisor (GCD) of two numbers 
# using the Euclidean algorithm.

# Algorithm:

# Divide the larger number by the smaller number
# Replace the larger number with the smaller number
# Replace the smaller number with the remainder
# Repeat until the remainder is 0
# The GCD is the last non-zero remainder
# Hint: While b != 0: store remainder (a%b), then a=b, 
# b=remainder. When loop ends, a is the GCD.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b!=0:
    r=a%b
    a=b
    b=r

    print(a)

```

```python
# Problem 6.3: Perfect Number Checker (Hard)
# Write a program to check if a number is a perfect number.

# A perfect number is a positive integer that is equal to the sum of its 
# proper positive divisors (excluding the number itself).

x=int(input("enter a number"))

sum=0

for i in range (1,x):
      if x% i == 0:
        sum = sum + i

if sum==x:
   print("its a perfct num")
else:
    print("its not a perfct")
```

```python
# Problem 6.4: LCM Calculator (Medium)
# Write a program to find the Least Common Multiple (LCM) of two numbers.

# Hint: LCM(a, b) = (a × b) / GCD(a, b)

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

a=x
b=y


while b!=0:
    r=a%b
    a=b
    b=r
    

gcd=a

lcm=(a*b)//gcd
```
