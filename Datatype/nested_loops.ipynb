```python
# Problem 4.1: Pattern Printing - Right Triangle (Easy)

n=5
for i in range (1,n+1):
    for j in range(1,i+1):
        print("*",end=" ")
    print()
```

```python
# Problem 4.2: Pattern Printing - Inverted Right Triangle (Easy)

n=5
for i in range (n,0,-1):
    for j in range (1,i+1):
            print("*", end=" ")
    print()
```

```python
# Problem 4.3: Pattern Printing - Pyramid (Medium)
n=10
for i in range(1,n+1):
        print(" " * (n - i), end="")   # print spaces
        print("*" * (2 * i - 1))       # print stars
```

```python
# Problem 4.4: Pattern Printing - Number Pattern (Medium)
a=5
for i in range(1,a+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print(" ")


```

```python
# Problem 4.5: Multiplication Table Grid 

a=5

for i in range(1,a+1):
    for j in range(1,a+1):
        print(i*j,end=" ")
    print()
```

```python
# Problem 4.6: Diamond Pattern 

n=5

for i in range(1,n+1):
    print(" "*(n-i)+"*"*(2*i-1))

for i in range(n-1, 0, -1):
    print(" "*(n-i) + "*"*(2*i-1))
```
